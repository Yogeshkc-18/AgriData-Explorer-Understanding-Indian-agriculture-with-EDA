```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset Load 

df = pd.read_csv("C:/Users/Yokesh/AgriData-Explorer-Understanding-Indian-agriculture-with-EDA/ICRISAT-District Level Data - ICRISAT-District Level Data.csv")

# Dataset Structure 
print("dataset info")
print(df.info())

# Columns & Data Types 
print("\n Columns List:")
print(df.columns)

# Basic Statistics (Summary)
print("\n Basic Statistics:")
print(df.describe())

# Missing Values Count
print("\n Missing Values Count:")
print(df.isnull().sum())

# Zero Values Count 
print("\n Zero Values Count:")
print((df == 0).sum())

#  Bar Chart: Top 7 Rice Producing States
top_rice = df.groupby('State Name')['RICE PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).head(7)
plt.figure(figsize=(10,6))
sns.barplot(x=top_rice.values, y=top_rice.index, palette='viridis')
plt.title("Top 7 Rice Producing States")
plt.xlabel("Total Rice Production (1000 tons)")
plt.ylabel("State")
plt.show()

#  Line Chart: India's Sugarcane Production Over Last 50 Years
sugarcane = df.groupby('Year')['SUGARCANE PRODUCTION (1000 tons)'].sum()
plt.figure(figsize=(10,6))
plt.plot(sugarcane.index, sugarcane.values, marker='o', color='orange')
plt.title("India's Sugarcane Production Over Last 50 Years")
plt.xlabel("Year")
plt.ylabel("Sugarcane Production (1000 tons)")
plt.grid(True)
plt.show()

#  Scatter Plot: Rice Area vs Production
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='RICE AREA (1000 ha)', y='RICE PRODUCTION (1000 tons)', hue='State Name', legend=False)
plt.title("Rice Area vs Production (By Districts)")
plt.xlabel("Rice Area (1000 ha)")
plt.ylabel("Rice Production (1000 tons)")
plt.show()

#  Correlation Heatmap (Numeric Columns)
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Between Crops Data")
plt.show()

print(" EDA Completed Successfully!")
```
